## XG

Faça uma visualização para associar o XG acumulado de cada jogador por sua quantidade de chutes na competição.

```python
# Pegar dados de eventos da competição

eventos_copa_america24 = sb.competition_events(
     country='South America',
     division='Copa America',
     season='2024',
     gender="male")

eventos_copa_america24.head(3)
```

```python
xg_por_chute = eventos_copa_america24.loc[:,['shot_statsbomb_xg','player','team']]
xg_por_chute = xg_por_chute.dropna(subset=['shot_statsbomb_xg'])
xg_por_chute
```

```python
# Contagem de chutes por atleta
qtd_chutes = xg_por_chute.player.value_counts()
qtd_chutes
```

```python
# xg acumulado por jogador
xg_acumulado = xg_por_chute.groupby('player', as_index=False)['shot_statsbomb_xg'].sum()
xg_acumulado= xg_acumulado.sort_values(by='shot_statsbomb_xg', ascending=False)
xg_acumulado
```

```python
chutes_xg = xg_acumulado.merge(qtd_chutes, on='player')
chutes_xg
```

## xG Acumulado vs Números de chutes por jogador

Esse gráfico mostra a relação entre a quantidade de chutes realizados por cada jogador e seu xG (gols esperados) acumulado ao longo da competição.

Permite identificar os jogadores que tiveram mais chances de gol comparado ao número de chutes na competição.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

chutes_xg_filtro = chutes_xg[(chutes_xg['count'] >= 5) | (chutes_xg['shot_statsbomb_xg'] >= 1.5)]

scatter = plt.scatter(
    chutes_xg['count'],
    chutes_xg['shot_statsbomb_xg'],
    cmap = 'plasma',
    edgecolors='black'
)

for i in range(len(chutes_xg_filtro)):
  plt.text(
      chutes_xg_filtro['count'].iloc[i] + 0.2,
      chutes_xg_filtro['shot_statsbomb_xg'].iloc[i],
      chutes_xg_filtro['player'].iloc[i],
      fontsize=9
  )

plt.xlabel('Quantidade de chutes')
plt.ylabel('xG Acumulado')
plt.title('xG Acumulado vs Número de chutes por jogador')
plt.show()
```

## Mapa de chutes do Messi - Copa América

O mapa representa todos os chutes realizados por Lionel Messi durante a Copa América, indicando suas posições no campo, o valor de xG (com tamanho dos pontos) e se o chute foi gol ou não (com formatos diferentes).

Esse gráfico ajuda a entender de onde o Messi mais finaliza e quais posições essa finalização pode representar mais perigo.

```python
from mplsoccer import VerticalPitch
import matplotlib.pyplot as plt

# Extrai os dados
x = chutes_messi['location'].apply(lambda loc: loc[0])
y = chutes_messi['location'].apply(lambda loc: loc[1])
xg = chutes_messi['shot_statsbomb_xg']
resultado = chutes_messi['shot_outcome']

# Resultado de cada chute
gols = resultado == 'Goal'
sot = resultado == 'Saved'
fora = resultado == 'Off T'
bloqueado = resultado == 'Blocked'
trave = resultado == 'Post'

gols_total = len(chutes_messi[chutes_messi['shot_outcome'] == 'Goal'])
xg_messi = xg.sum()
total_chutes = len(chutes_messi)

# Cria o campo
pitch = VerticalPitch(pitch_color='#87CEFA', half=True, line_color='white')
fig, ax = pitch.draw(figsize=(12, 10))

# Plota gols
scatter1 = pitch.scatter(x[gols], y[gols], s=xg[gols] * 1000, c='white', cmap='Reds', edgecolors='black', marker='football', ax=ax, label='Gol')

#Plota chutes ao gol
scatter2 = pitch.scatter(x[sot], y[sot], s=xg[sot] * 1000, c='yellow', cmap='Reds', edgecolors='black', ax=ax, label='Chute ao gol', marker='o')

#Plota chutes bloqueados
scatter3 = pitch.scatter(x[bloqueado], y[bloqueado], s=xg[bloqueado] * 1000, c='yellow', cmap='Reds', edgecolors='black', ax=ax, label='Chute bloqueado', marker='s')

#Plota chutes para fora
scatter4 = pitch.scatter(x[fora], y[fora], s=xg[fora] * 1000, c='yellow', cmap='Reds', edgecolors='black', ax=ax, label='Chute para fora', marker='^')

#Plota chutes na trave
scatter5 = pitch.scatter(x[trave], y[trave], s=xg[trave] * 1000, c='yellow', cmap='Reds', edgecolors='black', ax=ax, label='Chute na trave', marker='p')

plt.title('Mapa de Chutes do Messi - Copa América', fontsize=20)
txt = ax.text(x=5, y=70, s=f'Gols = {gols_total}\nxG = {xg_messi:.2f}\nChutes = {total_chutes}',
              size=18,
              color='white',
              va='center', ha='left')
plt.legend()
plt.show()
```

# XG Timeline

Faça uma timeline dos gols esperados de cada time na final da Copa América entre Colômbia e Argentina

```python
finalCopa = sb.events(match_id=3943077)
finalCopa
```

```python
chutes_final = finalCopa.loc[(finalCopa.type == 'Shot') ]
chutes_final
```

```python
chutes_final.columns[65:82]
```

```python
chutes_final_re = chutes_final[['minute','team','shot_statsbomb_xg','player','shot_outcome']]
chutes_final_re
```

```python
chutes_final_re = chutes_final_re.sort_values (by='minute')

chutes_final_re['xg_acumulado'] = chutes_final_re.groupby('team')['shot_statsbomb_xg'].cumsum()
chutes_final_re
```

## Timeline de gols esperados - Colômbia x Argentina

A timeline exibe, ao longo do tempo de jogo (em minutos), o valor de xG gerado em cada finalização. Também marca os momentos em que ocorreram gols mostrando o nome do jogador que marcou.

Esse gráfico ajuda a entender os momentos e evolução de cada equipe na partida.

```python
plt.figure(figsize=(10,7))

for team in chutes_final_re['team'].unique():
  dados_time = chutes_final_re[chutes_final_re['team']== team]
  plt.step(dados_time['minute'], dados_time['xg_acumulado'], where='post', label=team)

  gols=dados_time[dados_time['shot_outcome']=='Goal']
  plt.scatter(gols['minute'], gols['xg_acumulado'], edgecolors='black', s=50, zorder=5)

  for i in range(len(gols)):
    plt.text(
        gols['minute'].iloc[i] + 0.2,
        gols['xg_acumulado'].iloc[i],
        gols['player'].iloc[i],
        fontsize=9,
        ha='left',
        va='bottom'
    )

plt.xlabel('Minuto do Jogo')
plt.ylabel('xG Acumulado')
plt.title('Timeline de xG - Colômbia x Argentina')
plt.legend()
plt.show()
```
